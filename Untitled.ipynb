```python
BRs = [0.1,0.3,0.5,0.1]
sum(BRs)
cuts = [0]
for x in BRs:
    cuts.append(cuts[-1]+x)
cuts
```

```python
cats = []
for i in np.random.random(size = 100000):
    rand = np.random.random()
    cat = sum([1 if rand > x else 0 for x in cuts])
    cats.append(cat)
```

```python
_=plt.hist(cats)
```
